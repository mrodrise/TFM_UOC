```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
def grafica_rec(fichero):
    training_rewards = np.loadtxt('./resultados_ejecuciones/T2_QVC_DQN_CartPole_param2_tr.txt', dtype=float)
    mean_training_rewards = np.loadtxt('./resultados_ejecuciones/T2_QVC_DQN_CartPole_param2_mean_tr.txt', dtype=float)
    training_loss = np.loadtxt('./resultados_ejecuciones/T2_QVC_DQN_CartPole_param2_loss.txt', dtype=float)
    coefficients, residuals, _, _, _ = np.polyfit(range(len(mean_training_rewards)),mean_training_rewards, 1, full=True)
    plt.figure(figsize=(12, 8))
    plt.plot(training_rewards, label='Recompensas')
    plt.plot(mean_training_rewards, label='Recompensas medias')
    plt.axhline(200, color='r', label="Límite recompensa")
    plt.plot([coefficients[0] * x + coefficients[1] for x in range(len(mean_training_rewards))], label="Tendencia recompensa media")
    plt.xlabel('Episodios')
    plt.ylabel('Recompensas')
    plt.legend(loc="upper left")
    plt.show()
```

```python
def grafica_comparativa(lista):

    # Obtener el generador cíclico de estilos
    colors = [mpl.rcParams['axes.prop_cycle'].by_key()['color']][0]
    plt.figure(figsize=(12, 8))
    i=0
    for l in lista:
        # Obtener el primer estilo de la secuencia (que incluye el color predeterminado)
        color_predeterminado = colors[i]
        i += 1
        print(color_predeterminado)
        nombre, fichero = l
        plt.plot(np.loadtxt('./resultados_ejecuciones/'+fichero+"_mean_tr.txt", dtype=float), label=nombre)
        plt.plot(np.loadtxt('./resultados_ejecuciones/'+fichero+"_tr.txt", dtype=float), alpha=0.25, color=color_predeterminado)
    plt.axhline(195, color='r', label="Límite recompensa")
    plt.xlabel('Episodios')
    plt.ylabel('Recompensas')
    plt.legend(loc="lower right")
    plt.show()
```

```python
lista1 = [["NN DQN CartPole 1", "T2_Classic_DQN_CartPole_param1"], ["NN DQN CartPole 2", "T2_Classic_DQN_CartPole_param2"]]
lista2 = [["QVC DQN CartPole 1-a", "T2_QVC_DQN_CartPole_param1_a"], ["QVC DQN CartPole 1-b", "T2_QVC_DQN_CartPole_param1_b"]]
lista3 = [["QVC DQN CartPole 2-a", "T2_QVC_DQN_CartPole_param2_a"], ["QVC DQN CartPole 2-b", "T2_QVC_DQN_CartPole_param2_b"]]
lista4 = [["QVC DQN CartPole 2-b con lr 0.01", "T2_QVC_DQN_CartPole_param2_b_001"], ["QVC DQN CartPole 2-b con lr 0.01 y 3 capas","T2_QVC_DQN_CartPole_param2_b_001_3capas"]]
grafica_comparativa(lista1)
grafica_comparativa(lista2)
grafica_comparativa(lista3)
grafica_comparativa(lista4)
```
